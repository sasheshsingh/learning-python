```python
## Python Calculator

operator = input("Enter an operator (+, -, *, /): ")
num1 = int(input("Enter the 1st number: "))
num2 = int(input("Enter the 2nd number:"))
ans = None
if operator == "+":
    ans = num1 + num2
elif operator == "-":
    ans = num1 - num2
elif operator == "*":
    ans = num1 * num2
elif operator == "/":
    ans = num1 / num2
else:
    print("Not a valid operator")
print(round(ans))
```

```python
## Python Weight Calculator

weight = float(input("Enter your weight!"))
unit = input("Kilograms or Pounds? (K or L): ")

if unit == "K":
    weight = weight * 2.205
    unit = "Lbs."
    print(f"Your Weight is {round(weight, 2)} {unit}")
elif unit == "L":
    weight = weight / 2.205
    unit = "Kgs."
    print(f"Your Weight is {round(weight, 2)} {unit}")
else:
    print(f"{unit} was not valid.")

```

```python
unit = input("Is this temperature in Celcius or Fahrenheit (C/F): ")
temp = float(input("Enter the temperature"))

if unit == "C":
    temp = round((9 * temp) / 5  + 32, 1)
    print(f"The temperature in Farhenheit is: {temp} F")
elif unit == "F":
    temp = (temp - 32) * 5 / 9
    print(f"The temperature in Celcius is: {temp} C")
else:
    print(f"{unit} is an invalid unit of measurement")
```

## while loop

```python
name = input("Enter you name: ")
if name == "":
    print("You did not enter your name")
else:
    print(f"Hello {name}")
```

```python
name = input("Enter you name: ")
while name == "":
    print("You did not enter your name")
    name = input("Enter you name: ")
    if name != "":
        break
```

```python
age = int(input("Enter your age: "))
while age <= 0:
    print("Age can not be negative")
    age = int(input("Enter your age: "))
print(f"You are {age} years old")
```

```python
food = input("Enter a food you like (q to quit): ")
while not food == "q":
    print(f"You like {food}")
    food = input("Enter a food you like (q to quit): ")
```

```python
## Compound Interest Calculator
principle = 0
rate = 0
time = 0
while principle<=0:
    principle = float(input("Enter your principle amount: "))
    if principle <= 0:
        print("Principle cannot be less than or equal to zero")

print(principle)

while rate <= 0:
    rate = float(input("Enter your rate amount: "))
    if rate <= 0:
        print("Rate cannot be less than or equal to zero")

while time <= 0:
    time = float(input("Enter your time amount: "))
    if time <= 0:
        print("Time cannot be less than or equal to zero")


print(principle)
print(rate)
print(time)
```

```python
total = principle  * pow((1 + rate / 100), time)

print(f"Balance after {time} year/s: $ {total:.2f}")
```
