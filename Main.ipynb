# TypeCasting =  The Process of converting a from one data type to another
## Ex: str(), int(), float(), bool()

```python
name = "Sashesh Singh"
age = 25
gpa = 7.9
is_student = False
```

```python
print("name: ", type(name))
print("age: ", type(age))
print("gpa: ", type(gpa))
print("is_student: ", type(is_student))
```

```python
print("type of age: ", age, type(age))
age = float(age)
print("type of age: ", age, type(age))
age += 1
print("type of age: ", age, type(age))
age = str(age)
print("type of age: ", age, type(age))
```

```python
age += "1"
print("type of age: ", age, type(age))
```

```python
age = bool(age)
print("type of age: ", age, type(age))
age = 0
age = bool(age)
print("type of age: ", age, type(age))
```
